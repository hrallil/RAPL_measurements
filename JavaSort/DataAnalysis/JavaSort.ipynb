```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("JavaSort.csv")
df.columns = ["Name","PKG","CPU", "GPU" , "DRAM?" , "Time (sec)"]
Q1 = df[df["Name"] == "JavaSort1"]
Q2 = df[df["Name"] == "JavaSort2"]
Q3 = df[df["Name"] == "JavaSort3"]
Q4 = df[df["Name"] == "JavaSort4"]
Q5 = df[df["Name"] == "JavaSort5"]
Q6 = df[df["Name"] == "JavaSort6"]
Q7 = df[df["Name"] == "JavaSort7"]
Q8 = df[df["Name"] == "JavaSort8"]
Q9 = df[df["Name"] == "JavaSort9"]
Q10 = df[df["Name"] == "JavaSort10"]

print(df)

d = {"100" : Q1["Time (sec)"].mean(), 
     "200" : Q2["Time (sec)"].mean(),
     "300" : Q3["Time (sec)"].mean(), 
     "400" : Q4["Time (sec)"].mean(),
     "500" : Q5["Time (sec)"].mean(), 
     "600" : Q6["Time (sec)"].mean(),
     "700" : Q7["Time (sec)"].mean(), 
     "800" : Q8["Time (sec)"].mean(),
     "900" : Q9["Time (sec)"].mean(), 
     "1000" : Q10["Time (sec)"].mean()
    }
small = pd.Series(data = d)


def plotLinReg(x,y,dataFrame, title):
    variance = x.var() # The variance of x
    covariance = x.cov(y) # The covariance between x and y

    print(f'The variance is {variance}, and the covariance is {covariance}')
    a = x.cov(y)/x.var()

    print(f'The slope is a = {a}')
    b = y.mean()-a*x.mean()

    print(f'The intersection is b = {b}')
    y_fit = a*x + b
    dataFrame['fit'] = y_fit

    dataFrame.plot.scatter(0, 1, figsize=(8, 6), color="red") # Plot column index 0 against column index 
    plt.plot(x, y_fit, '-')
    plt.title(title)
```

```python
plt.plot(df["Time (sec)"],".")
plt.ylabel('time (ms)')
plt.xlabel('Array size')
plt.title("Small - time ")
```
