```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfMB = pd.read_csv("Idle_2.csv") # not generated yet
```

```python
def meanTimePrDataSize(df, name, sizes):
    df.columns = ["Name","PKG","CPU", "GPU" , "DRAM?" , "Time (sec)"]
    d = {}
    for i in range(len(sizes)):
        d[str(sizes[i])] = df[df["Name"] == (name+str(sizes[i]))] ["Time (sec)"].mean()
    
    return pd.Series(data = d)
```

```python
print("Example quicksort Data 25k - 1m")
sizes = [i*10000 for i in range(1,11)]
#print(sizes)
s = meanTimePrDataSize(dfMB, "quicksort", sizes)
print(s)
showGraphs(dfMB, s)
```
