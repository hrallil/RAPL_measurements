```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#load CSV
pc12 = pd.read_csv("pcTest-12.csv")
pc27 = pd.read_csv("pcTest-27.csv")
# pc 25 didnt want to pull the git repo

pc12.columns = ["Name","PKG","CPU", "GPU" , "DRAM?" , "Time (sec)"]
pc27.columns = ["Name","PKG","CPU", "GPU" , "DRAM?" , "Time (sec)"]

# delete any value < 0
pc12 = pc12[pc12['CPU'] >= 0]
pc27 = pc27[pc27['CPU'] >= 0]

#plot data
plt.plot(pc12["CPU"], ".",label="pc12")
plt.plot(pc27["CPU"], ".",label="pc27")

plt.legend(loc="upper left")
plt.xlabel("test number")
plt.ylabel("CPU (J)")
plt.title("all measurements")
plt.show()
```
