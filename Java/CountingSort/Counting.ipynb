Calculate mean time for each array:

    s = meanTimePrDataSize(df, "name", sizes)

And plot the graphs:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfC = pd.read_csv("JavaC.csv")
dfC2 = pd.read_csv("countRepeat9-3.csv")
```

```python
def meanTimePrDataSize(df, name, sizes):
    df.columns = ["Name","PKG","CPU", "GPU" , "DRAM?" , "Time (sec)"]
    d = {}
    for i in range(len(sizes)):
        d[str(sizes[i])] = df[df["Name"] == (name+str(sizes[i]))] ["Time (sec)"].mean()
    
    return pd.Series(data = d)
```

```python
def showGraphs(df, s):
    sns.set(rc={'figure.figsize':(15,8.27)})

    plt.plot(df["Time (sec)"],".")
    plt.ylabel('time (ms)')
    plt.xlabel('Measurment NO')
    plt.show()
    
    plt.plot(df["CPU"],".")
    plt.ylabel('Energy (CPU) [J]')
    plt.xlabel('Measurement NO')
    plt.show()
    
    plt.plot(s, ".")
    plt.ylabel('Time (ms)')
    plt.xlabel('Array size')
    plt.xticks(rotation = 30)
    plt.show()
    
    box = sns.boxplot(data=df, x='Name', y = 'Time (sec)')
    box.set_xticklabels(box.get_xticklabels(),rotation=30)
```

```python
print("Counting Sort Data 100 - 100000")
sizes = [i*10*pow(10,j) for j in range(1,4) for i in range(1,10)]
sizes.append(100000)

s = meanTimePrDataSize(dfC, "countingSort", sizes)
#print(s)
showGraphs(dfC, s)
```

```python
print("Counting Sort repeat Data 100 - 100000")
#sizes = [i*10*pow(10,j) for j in range(1,4) for i in range(1,10)]
#sizes.append(100000)
s = meanTimePrDataSize(dfC2, "countingSort", sizes)
showGraphs(dfC2, s)
```
