```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Java.csv")
df.columns = ["Name","CPU", "GPU" , "DRAM?" , "Time (sec)", "?"]
INT1024 = df[df["Name"] == "bubblesortInt1024"]
INT512 = df[df["Name"] == "bubblesortInt512"]
LONG1024 = df[df["Name"] == "bubblesortLong1024"]
LONG512 = df[df["Name"] == "bubblesortLong512"]

FLOAT1024 = df[df["Name"] == "bubblesortFloat1024"]
FLOAT512 = df[df["Name"] == "bubblesortFloat512"]
DOUBLE1024 = df[df["Name"] == "bubblesortDouble1024"]
DOUBLE512 = df[df["Name"] == "bubblesortDouble512"]
print(INT1024)
```

```python

plt.plot(INT1024["?"],INT1024["CPU"],".",label="CPU")
plt.plot(INT1024["?"], INT1024["GPU"],".",label="GPU")
plt.legend()
plt.ylabel('Energy?')
plt.xlabel('time?')
plt.title("INT 1024")
plt.show()

plt.plot( LONG1024["?"],LONG1024["CPU"],".",label="CPU")
plt.plot( LONG1024["?"],LONG1024["GPU"],".",label="GPU")
plt.legend()
plt.ylabel('Energy?')
plt.xlabel('time?')
plt.title("LONG 1024")
plt.show()

plt.plot( FLOAT1024["?"],FLOAT1024["CPU"],".",label="CPU")
plt.plot( FLOAT1024["?"],FLOAT1024["GPU"],".",label="GPU")
plt.legend()
plt.ylabel('Energy?')
plt.xlabel('time?')
plt.title("FLOAT 1024")
plt.show()

plt.plot( DOUBLE1024["?"],DOUBLE1024["CPU"],".",label="CPU")
plt.plot( DOUBLE1024["?"],DOUBLE1024["GPU"],".",label="GPU")
plt.ylabel('Energy?')
plt.xlabel('time?')
plt.title("DOUBLE 1024")
plt.legend()
plt.show()
```

```python
plt.plot( INT1024["?"],INT1024["CPU"],".",label="INT CPU")
plt.plot( INT1024["?"],INT1024["GPU"],".",label="INT GPU")

plt.plot( LONG1024["?"],LONG1024["CPU"],".",label="LONG CPU")
plt.plot( LONG1024["?"],LONG1024["GPU"],".",label="LONG GPU")

plt.plot( FLOAT1024["?"],FLOAT1024["CPU"],".",label="FLOAT CPU")
plt.plot( FLOAT1024["?"],FLOAT1024["GPU"],".",label="FLOAT GPU")

plt.plot( DOUBLE1024["?"],DOUBLE1024["CPU"],".",label="DOUBLE CPU")
plt.plot( DOUBLE1024["?"],DOUBLE1024["GPU"],".",label="DOUBLE GPU")
plt.legend()
plt.title("All 1024")
plt.ylabel('Energy?')
plt.xlabel('time?')

```
