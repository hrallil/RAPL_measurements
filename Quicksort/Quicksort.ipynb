```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("Javaq.csv")
df.columns = ["Name","PKG","CPU", "GPU" , "DRAM?" , "Time (sec)"]
Q1 = df[df["Name"] == "quicksort1"]
Q2 = df[df["Name"] == "quicksort2"]
Q3 = df[df["Name"] == "quicksort3"]
Q4 = df[df["Name"] == "quicksort4"]
Q5 = df[df["Name"] == "quicksort5"]
Q6 = df[df["Name"] == "quicksort6"]
Q7 = df[df["Name"] == "quicksort7"]
Q8 = df[df["Name"] == "quicksort8"]
Q9 = df[df["Name"] == "quicksort9"]
Q10 = df[df["Name"] == "quicksort10"]

d = {"100" : Q1["Time (sec)"].mean(), 
     "200" : Q2["Time (sec)"].mean(),
     "300" : Q3["Time (sec)"].mean(), 
     "400" : Q4["Time (sec)"].mean(),
     "500" : Q5["Time (sec)"].mean(), 
     "600" : Q6["Time (sec)"].mean(),
     "700" : Q7["Time (sec)"].mean(), 
     "800" : Q8["Time (sec)"].mean(),
     "900" : Q9["Time (sec)"].mean(), 
     "1000" : Q10["Time (sec)"].mean()
    }
s = pd.Series(data = d)
s
```

```python
plt.plot(df["Time (sec)"],".")
plt.ylabel('time (ms)')
plt.xlabel('Array size')
plt.title("Small - time ")
```

```python
plt.plot(df["CPU"],".")
plt.ylabel('Energy (CPU)')
plt.xlabel('Array size')
plt.title("Small - energy")
```

```python
plt.plot(s, ".")
plt.ylabel('Time (ms)')
plt.xlabel('Array size')
plt.title("Small - time (avg)")
```
