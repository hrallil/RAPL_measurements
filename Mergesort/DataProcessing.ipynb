```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
dfMB = pd.read_csv("JavaM.csv")
dfMS = pd.read_csv("JavaMs.csv")
#dfC = pd.read_CSV("JavaC.csv")
```

```python
def meanTimePrDataSize(df, name, sizes):
    df.columns = ["Name","PKG","CPU", "GPU" , "DRAM?" , "Time (sec)"]
    d = {}
    for i in range(len(sizes)):
        d[str(sizes[i])] = df[df["Name"] == (name+str(i))]["Time (sec)"].mean()
    
    return pd.Series(data = d)
```

```python
def showGraphs(df, s):
    plt.plot(df["Time (sec)"],".")
    plt.ylabel('time (ms)')
    plt.xlabel('Measurment NO')
    plt.show()
    
    plt.plot(df["CPU"],".")
    plt.ylabel('Energy (CPU) [J]')
    plt.xlabel('Measurement NO')
    plt.show()
    
    plt.plot(s, ".")
    plt.ylabel('Time (ms)')
    plt.xlabel('Array size')
    plt.show()
```

```python
print("Merge sort Data 10000 - 100000")
sizes = [10000,20000,30000,40000,50000,60000,70000,80000,90000,100000]
s = meanTimePrDataSize(dfMB, "merge-sort", sizes)
showGraphs(dfMB, s)
```

```python
print("Merge Sort Data 100 - 1000")
sizes = [100,200,300,400,500,600,700,800,900,1000]
s = meanTimePrDataSize(dfMS, "merge-sort", sizes)

showGraphs(dfMS, s)
```
